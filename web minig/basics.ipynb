```python
import requests #get webpage data as an object
from bs4 import BeautifulSoup #parse(selection) webpage data 
```

```python
try:
    url = 'https://blog.jetbrains.com/kotlin/'
    page = requests.get(url)
    if page.status_code == 200:
        print(f'{page.status_code} Sucess!')
        soup = BeautifulSoup(page.content, 'html.parser')# this line is used to fetch data
    elif page.status_code == 404:
        print(f'{page.status_code}page not found')
    elif page.status_code == 403:
        print(f'{page.status_code}Forbidden')
    elif page.status_code == 500:
        print(f'{page.status_code}Internal service Error')
    else:
        print(f'{page.status_code}Unknown Error')
except Exception as e:
    print(f'Error: \n{e}')
```

```python
target = soup.find('div', class_ = 'row latest latest_posts_section')#class_ is propety of find to find thath property
if target:
    print('target section found!')
```

```python
articles = target.find_all('div', class_ = 'col')
if articles:
    print('Articles found!')
    print(f'total articles: {len(articles)}')
```

```python
for item in articles:
    heading  = item.find('h3')
    publish = item.find('time')
    summary = item.find('p')
    author = item.find('span')
    print('-'*20)
    print(heading.text)
    print(publish['datetime'])
    print(summary.text)
    print(author.text)
```
