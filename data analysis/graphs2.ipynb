```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

# categorical graphs
data is said to be categorical when the values of that set are less and specific e.g in titanic survived and pclass is categorical
- columns that have text generally
- first count unique values
- plot the count of each category
- following charts are a good way to visualiz data
    - bar chart
    - pie chart
    - line chart 
    - donut chart
    - area chart

```python
print(titanic['survived'].unique())
print(titanic['pclass'].unique())
print(titanic['age'].unique())
```

```python
survival_count = titanic['survived'].value_counts()
pclass_k = titanic['pclass'].value_counts()

survival_count.plot(kind='pie', autopct= '%.1f%%', title= 'survival rate')#ye shortcut hai plot karne ka isi ko use kare
```

```python
survival_count.plot(
    kind = 'pie',
    autopct = '%.1f%%',
    title = 'survival rate',
    labels= ['survived','not survived'],
    colors= ['#66b3ff','#ff9999'],
    labeldistance = 1.1,
    wedgeprops = {
       'width':0.5#agar donut banana hai to wedgeprop mei width humesha 0 se 1 ke "beech" mei rakho
    }
)
```

```python
pclass_k.plot(
    kind = 'pie',
    autopct = '%.1f%%',
    title = 'number of ppl in deck x',
    labels= ['3','1','2'],
    colors= ['#66b3ff','#ff9999','#abc'],
    labeldistance = 1.1,
    wedgeprops = {
       'width':0.5#agar donut banana hai to wedgeprop mei width humesha 0 se 1 ke "beech" mei rakho
    }
)
```
