```python
penguins = sns.load_dataset('penguins')

```

```python
def categorize_by_w(value):
    if value < np.nan:
         return np.nan
    elif value <3900:
         return 'light weight'
    elif value > 3900 and value <5100:
         return 'medium weight'
    elif value>5100:
         return 'heavy weight'
penguins['category'] = penguins['body_mass_g'].apply(categorize_by_w)
penguins
    
```

```python
penguins
```

```python
penguins.groupby('species')[['bill_length_mm','body_mass_g']].max()
```

```python
penguins.groupby('island')[['bill_length_mm','body_mass_g']].max().plot(kind = 'bar')
```

```python
penguins.groupby(['species','island','sex'])[['bill_length_mm','body_mass_g']].max()
```

```python
penguins.groupby(['species','island','sex'])[['bill_length_mm','body_mass_g']].max().reset_index()#reset index bass unn khli spaces ko appropriately bhar deta hai
```

```python
penguins.pivot_table(values='body_mass_g', index= 'species',columns= 'island')#by default pivot permorms avg function if you eant another functions write aggfunc in the brackets
```

```python
penguins.pivot_table(values='body_mass_g', index= 'species',columns= 'island',aggfunc='max').plot(kind = 'bar')
```

```python
pp= penguins.pivot_table(values='body_mass_g', index= 'species',columns= 'island',aggfunc='max')
pp.replace(np.nan, 0, inplace = True)
pp = pp.astype(int)
pp
```
